## Линейная регрессия

Если хотим предсказать значение НЕПРЕРЫВНОЙ ПЕРЕМЕННОЙ - решаем задачу регрессии

```python
#сразу практика
import pandas as pd
laptops=pd.read_csv('data/laptop_price.csv',encoding='Windows-1251')
laptops.head()
```

```python
laptops.info()
```

Результирующая переменная - Price_euros. Будем предсказывать цену ноутбука по его характеристикам.

```python
#попробуем предсказать значение цены по весу ноутбука
#
new_weight=[]
for i in laptops['Weight']:
    new_weight.append(float(i.replace('kg','')))
laptops['Weight']=new_weight
laptops.head()
```

```python
import plotly.express as px
px.scatter(laptops,x='Weight',y='Price_euros')
```

## Реализация линейной регрессии в scikit-learn

```python
#для начала предобратаем данные
#на основе Memory выделим три признака по типу памяти SSD, HDD, Flash Storage
ssd=[]
hdd=[]
fs=[]
for i in laptops['Memory']:
    if 'SSD' in i:
        ssd.append(1)
        hdd.append(0)
        fs.append(0)
    elif 'HDD' in i:
        ssd.append(0)
        hdd.append(1)
        fs.append(0)
    else:
        ssd.append(0)
        hdd.append(0)
        fs.append(1)
laptops['ssd']=ssd
laptops['hdd']=hdd
laptops['fs']=fs
laptops.head()
```

```python
#от памяти оставим только объем
import re
s = [float(re.findall(r'-?\d+\.?\d*', s)[0]) for s in laptops['Memory']]
laptops['Memory']=s
laptops.head()
```

```python
laptops.columns
```

```python
laptops.drop(['laptop_ID','Product','ScreenResolution','Cpu','Gpu',],axis=1,inplace=True)
laptops.head()
```

```python
#переводим категориальные признаки в вектора
laptops=pd.get_dummies(laptops,columns=['Company','TypeName','Ram','OpSys'])
laptops.head()
```

```python
laptops.info()
```

```python
#делим на на признаки и результирующиюся
X=laptops.drop(['Price_euros'],axis=1)
Y=laptops['Price_euros']
```

```python
#разбиение на обучающую и тестовую выборку
#test_size - объем обучающей выборки
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

* Обучающая выборка - на ней обучаем нашу модель
* Тестовая - проверяем, как хорошо наша модель обучилась

```python
from sklearn.linear_model import LinearRegression #импорт для создания модели
regressor = LinearRegression() # а вот и наша модель
regressor.fit(X_train, y_train) # А тут обучение на обучающей выборке
```

```python
#свободный коэф
print(regressor.intercept_)
```

```python
print(regressor.coef_)
```

## Как проверить качество нашей модели?

```python
#для начала предскажем с помощью нашей модели цену для тестовых данных
y_pred=regressor.predict(X_test)
y_pred[:10]
```

```python
true_vs_pred = pd.DataFrame({'Истинные': y_test, 'Предсказанные': y_pred})
true_vs_pred.head()
```

```python
import plotly.graph_objects as go
import numpy as np

random_x = np.linspace(0, len(true_vs_pred['Истинные']))

fig = go.Figure()
fig.add_trace(go.Scatter(x=random_x, y=true_vs_pred['Истинные'],
                    mode='lines',
                    name='lines'))
fig.add_trace(go.Scatter(x=random_x, y=true_vs_pred['Предсказанные'],
                    mode='lines+markers',
                    name='lines+markers'))

fig.show()
```

## Метрики оценки качества регрессионной модели

Модель это хорошо, а как понять что она годная?

Статья про метрики - https://neerc.ifmo.ru/wiki/index.php?title=Оценка_качества_в_задачах_классификации_и_регрессии

```python
#коэффициент детерминации
from sklearn.metrics import r2_score
r2 = r2_score(y_test,y_pred)
print('Коэффициент детерминации:',r2)
```

```python
from sklearn import metrics 
print('Средняя абсолютная ошибка:', metrics.mean_absolute_error(y_test, y_pred)) 
print('Средняя квадратическая ошибка:', metrics.mean_squared_error(y_test, y_pred)) 
print('Корень из среднеквадратической ошибки', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

## Кросс-валидация

Кросс-валидация — это методика обучения и оценки модели, которая разбивает данные на несколько секций и обучает несколько алгоритмов на этих секциях.

```python
Image(url='https://wiki.loginom.ru/images/cross-validation.svg')
```

```python
from sklearn.model_selection import cross_validate #для кросс-валидации
#параметры - модель,Х,Y, сv(сколько разбиений),scoring - метрики оценки
scores = cross_validate(regressor, X, Y, cv=5,scoring=('r2'))
```

```python
scores
```

```python
scores['test_score'].mean().round(2) #среднее значение метрики на кросс-валидации
```
